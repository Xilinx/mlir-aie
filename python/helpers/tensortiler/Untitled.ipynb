```python
tiles = TensorTiler2D.simple_tiler((4, 4), (2, 2), tile_col_major=True, iter_col_major=True, tile_repeat=2)
for t in tiles:
    t.visualize(plot_access_count=True)
tiles.visualize(show_plot=True, plot_access_count=True)
```

```python
anim = tiles.animate(animate_access_count=True)
HTML(anim.to_jshtml())
```

```python
# tensor_dims, tile_dims, tile_group_dims, tile_col_major, tile_group_col_major, iter_col_major, tile_repeat, allow_partial
tiles = TensorTiler2D.group_tiler((16, 16), (2, 2), (2, 2), tile_col_major=True, tile_group_col_major=False, iter_col_major=False, tile_repeat=2)
anim2 = tiles.animate()
for t in tiles:
    print(t)
tiles[0].visualize(plot_access_count=True)
HTML(anim2.to_jshtml())
```
